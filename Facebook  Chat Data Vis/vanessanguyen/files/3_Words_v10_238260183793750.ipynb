```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
print ('Welcome to the Word Cluster Creator!')
print ('This visual representation of words distinguishes between 3 dimensions. Thoughtfulness, which represent intentionality, creativity, intelligence or analytical tendencies. Emotion, and Intesnsity. Intensity is a unique attribute in that it is a modifier for the other two attributes. For example, prolific would be a high intensity high thoughtfulness, while easy going would be low intensity low thoughfulness. Intensity modifies emotion in a unique way, the two ends of the spectrum would be resilience and fragility, so turbulent would be high emotion, low intensity.')
print ('Please enter how each attribute applies to each word on a scale from 0 - 10, with 5 being neutral.')
Str = input('Enter the words here, separated by commas with no spaces: ')
Array = Str.lower().split(',')
Array = set(Array)
wordsDF = pd.DataFrame(index=Array, columns=['Thoughtfulness', 'Emotion', 'Intensity'])
```

```python
for word in wordsDF.index.tolist():
    for attribute in wordsDF.columns.tolist():
        value = input('Enter ' + attribute + ' of the word ' + word + ': ')
        while not value.isdigit() or int(value) < 0 or int(value) > 10:
            value = input('Please enter a value between 0 and 10: ')
        wordsDF.loc[word][attribute] = int(value)
        
wordsDF
wordsDF.to_csv('/Users/ZachIp/Desktop/untitled folder/' + input("Save your plot") + '.csv')
        
```

```python
visual = plt.figure().gca(projection='3d')
visual.scatter(wordsDF.Thoughtfulness, wordsDF.Emotion, wordsDF.Intensity)
visual.set_xlabel('Thoughtfulness')
visual.set_ylabel('Emotion')
visual.set_zlabel('Intensity')
```
